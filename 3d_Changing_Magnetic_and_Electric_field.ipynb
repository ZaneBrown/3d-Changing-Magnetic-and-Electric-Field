```python
#cartesian
arr_position=[]
position=[0,0,0]
velocity=[0,0,0]
acceleration=[0,0,0]
E_Field=[0,0,0]
B_Field=[0,0,0]
t=0
dt=0.01
q=(1)

def cross_product(a, b):
  c = [a[1]*b[2] - a[2]*b[1], a[2]*b[0] - a[0]*b[2], a[0]*b[1] - a[1]*b[0]]
  return c

def lorentz_force():
  c=cross_product(velocity, B_Field)
  sum_F=[q*(E_Field[0]+c[0]), q*(E_Field[1]+c[1]), q*(E_Field[2]+c[2])]
  return sum_F

def magnetic_function(position_B,B_Func):
  B=[0,0,0]
  x,y,z=position_B[0:]
  for i in range(3):
    B[i] = eval(B_Func[i], {"x": x, "y": y, "z": z, "math": math})
  return B

def electric_function(position_E,E_Func):
  E=[0,0,0]
  x,y,z=position_E[0:]
  for i in range(3):
    E[i] = eval(E_Func[i], {"x": x, "y": y, "z": z, "math": math})
  return E


vel=[1500.0,2000.0,6000.0]
E_Field=[100.0,10.0,20.0]
B_Field=[88.0,165.0,600.0]
#B_Function=["math.cos(x)", "math.sin(y)", "z"]
#E_Function=["math.sin(y)", "-1*math.cos(x)", "-1*z"]
B_Function=["math.cos(x)", "math.sin(y)", "z"]
E_Function=["-math.cos(x)", "-math.sin(y)", "-1*z"]


arr_position.append(position)
while t <= 50:
  force=lorentz_force()
  for i in range(3):
    acceleration[i]=force[i]/q
    velocity[i]+=acceleration[i]*dt
    position[i]+=velocity[i]*dt
  B_Field=magnetic_function(position,B_Function)
  E_Field=electric_function(position,E_Function)
  arr_position.append(copy.copy(position))
  t+=dt


x,y,z=zip(*arr_position)
fig = make_subplots(rows=1, cols=1, specs=[[{'type': 'scatter3d'}]])
scatter = go.Scatter3d(x=x,y=y,z=z,mode='markers',marker=dict(size=6, color=z, colorscale='Viridis', opacity=0.8))
fig.add_trace(scatter)
fig.update_layout(scene=dict(aspectmode='cube'))
fig.show()
```
